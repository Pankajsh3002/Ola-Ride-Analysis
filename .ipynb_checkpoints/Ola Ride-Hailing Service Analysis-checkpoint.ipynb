# Ola Ride-Hailing Service Analysis

## Project Goal:
The primary goal of this project is to analyze Ola's ride-hailing dataset to derive actionable insights that can help the business improve operational efficiency, enhance customer and driver satisfaction, and ultimately reduce the rate of non-successful rides.

### 1. Identify Key Operational Trends:
We will analyze the distribution of ride statuses (Successful, Cancelled, Incomplete) to understand the overall health of the service. We will also investigate ride volume over time (daily and weekly patterns) and the popularity of different vehicle types to understand demand and customer preferences.

### 2. Diagnose Cancellation and Service Failure Root Causes:
This is the core of our analysis. We will dive deep into why rides fail. By segmenting cancellations by who initiated them (customer or driver), we will analyze the specific reasons provided. This will help us pinpoint the most significant pain points in the booking process, such as drivers asking customers to cancel or frequent "personal issues" from drivers.

### 3. Formulate Data-Driven Recommendations:
Based on the patterns and problems identified, we will synthesize our findings into concrete, actionable recommendations. For example, if we find that a specific cancellation reason is prevalent, we will suggest a targeted intervention. The final output will not just be a set of charts, but a strategic brief that proposes solutions to improve service reliability and user retention.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
file_path=r"C:\Users\jiten\Documents\Ola-Data-Analysis-Project\ola_data.csv"

#Reading Csv file
df=pd.read_csv(file_path,sep=',')
```

```python
#First 5 rows of the dataframe
df.head()
```

```python
#Summary of the dataset
df.info()
```

```python
#Getting descriptive statistics for numerical columns
df.describe()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Converting the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df[['Date']].info()
```

## Data Cleaning

```python
#Droping Avg VTAT and AVG CTAT Column Because these column has many missing values
df.drop(['Avg VTAT', 'Avg CTAT'], axis=1, inplace=True)
print("Dropped Avg VTAT and Avg CTAT columns")
```

```python
#Filling Missing Numerical and Categorical Data
numerical_cols = ['Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']
df[numerical_cols] = df[numerical_cols].fillna(0)
```

```python
#Filling missing payment method with a placeholder
df['Payment Method'].fillna('Not Applicable', inplace=True)
```

```python
#Filling missing reasons with descriptive text
df['Reason for Cancelling by Customer'].fillna('Not Cancelled', inplace=True)
df['Reason for Cancelling by Driver'].fillna('Not Cancelled', inplace=True)
df['Incomplete Rides Reason'].fillna('Not Applicable', inplace=True)

print("Cleaned up reason columns")
```

```python
#Creating new columns for day of week and month
df['Day of Week'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month_name()

print("Created Day of Week and Month columns")
```

```python
#Checking summary of the dataset
df.info()
```

## Data Analysis with Pandas

### High-Level Overview: Booking Status

```python
#Calculating the count of each booking status
booking_status_counts = df['Booking Status'].value_counts()

#Calculating the percentage for each status
booking_status_percent = df['Booking Status'].value_counts(normalize=True) * 100

#Combining them into a single, summary DataFrame
booking_status_analysis = pd.DataFrame({'Count': booking_status_counts,'Percentage': booking_status_percent.round(2)})

print("----------------------------------- Booking Status Analysis -------------------------")
print(booking_status_analysis)
```

### Cancellation Deep-Dive

```python
#Total count of cancellations by each party
total_cancellations = pd.DataFrame({
    'Party': ['Customer', 'Driver'],
    'Total Cancellations': [
        df['Cancelled  by Customer'].sum(),
        df['Cancelled Rides by Driver'].sum()]
}).set_index('Party')

print("Total Cancellations by Party")
print(total_cancellations)

# Finding the most common cancellation reasons for drivers
driver_cancellation_reasons = df[df['Reason for Cancelling by Driver'] != 'Not Cancelled']
common_driver_reasons = driver_cancellation_reasons['Reason for Cancelling by Driver'].value_counts()

print("\n----------------------- Top Driver Cancellation Reasons------------------------")
print(common_driver_reasons)
```

### Vehicle Performance and Revenue Analysis

```python
# Group by vehicle type to analyze multiple metrics at once
vehicle_analysis = df.groupby('Vehicle Type').agg(
    Total_Rides=('Booking ID', 'count'),
    Successful_Rides=('Booking Value', lambda x: (x > 0).sum()),
    Average_Booking_Value=('Booking Value', 'mean'),
    Average_Ride_Distance=('Ride Distance', 'mean')
).round(2)

print("----------------------- Vehicle Performance Analysis ----------------------")
vehicle_analysis=vehicle_analysis.sort_values(by='Total_Rides', ascending=False)
vehicle_analysis
```

### Rides Count by Days Analysis

```python
rides_by_day = df['Day of Week'].value_counts()

# Defining the correct order of days
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Reindexing the series to sort the days correctly and  filling missing days with 0
rides_by_day = rides_by_day.reindex(days_order).fillna(0)

print("---------------------------- Ride Counts by Day of the Week --------------------")
print(rides_by_day)
```

#### Overall Booking Status

```python
# Using the booking_status_analysis DataFrame 
plt.figure(figsize=(8, 8))

# Creating the Pie Chart
plt.pie(
    booking_status_analysis['Count'],
    labels=booking_status_analysis.index,
    autopct='%1.1f%%', 
    startangle=140,
    colors=sns.color_palette('viridis', len(booking_status_analysis))
)

plt.title('Proportional Distribution of Booking Statuses', fontsize=16)
plt.ylabel('') 
plt.show()
```

### The Cancellation Story

#### Who Cancelled?

```python
# Using the total_cancellations DataFrame
plt.figure(figsize=(8, 8))

plt.pie(
    total_cancellations['Total Cancellations'],
    labels=total_cancellations.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set1', len(total_cancellations)),
    wedgeprops=dict(width=0.4) # This creates the donut hole
)

plt.title('Share of Cancellations: Driver vs. Customer', fontsize=16)
plt.show()
```

#### Why Did Drivers Cancel?

```python
# Using the common_driver_reasons 
plt.figure(figsize=(12, 7))

sns.barplot(
    x=common_driver_reasons.values,
    y=common_driver_reasons.index,
    palette='plasma',
    orient='h'
)

plt.title('Top Reasons for Ride Cancellation by Drivers', fontsize=16)
plt.xlabel('Number of Cancellations', fontsize=12)
plt.ylabel('Reason', fontsize=12)
plt.tight_layout()
plt.show()
```

## Vehicle Performance

#### Most Popular Vehicle Types

```python
# Using the vehicle_analysis DataFrame
vehicle_popularity = vehicle_analysis.sort_values(by='Total_Rides', ascending=False)
plt.figure(figsize=(10, 6))

sns.barplot(x=vehicle_popularity.index, y='Total_Rides', data=vehicle_popularity, palette='magma')

plt.title('Total Rides Booked by Vehicle Type', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Total Number of Rides', fontsize=12)
plt.show()
```

```python
# Using the vehicle_analysis DataFrame
vehicle_value = vehicle_analysis.sort_values(by='Average_Booking_Value', ascending=False)
plt.figure(figsize=(10, 6))

sns.barplot(x=vehicle_value.index, y='Average_Booking_Value', data=vehicle_value, palette='coolwarm')

plt.title('Average Booking Value per Successful Ride', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Booking Value ($)', fontsize=12)
plt.show()
```

#### Ride Demand Throughout the Week

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.lineplot(
    x=rides_by_day.index,
    y=rides_by_day.values,
    marker='o', 
    linestyle='-',
    color='b'
)

plt.title('Ride Demand Distribution by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Rides Booked', fontsize=12)
plt.grid(True)
plt.show()
```

### Customer Payment Method Preferences

```python
# Analzing Successfull Rides
payment_counts = df[df['Booking Status'] == 'Success']['Payment Method'].value_counts()

plt.figure(figsize=(8, 8))


seaborn_palette = sns.color_palette('deep')
# Creating the pie chart using the 'viridis' colormap
plt.pie(
    payment_counts,
    labels=payment_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=seaborn_palette
)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Adding title 
plt.title('Share of Payment Methods for Successful Rides', fontsize=16)
plt.axis('equal')  

plt.show()
```

# Summary of Findings and Actionable Recommendations

# Key Findings

### 1. High Rate of Service Failure  
A significant portion of booked rides do not complete successfully. Our analysis shows a majority of bookings in the sample were cancelled by drivers.  

### 2. Driver-Initiated Cancellations are the Primary Issue  
Drivers are the main source of cancellations. The top reasons point towards either external issues ("Personal & Car related issues") or potential platform manipulation ("The customer was coughing/sick" could be a way to avoid penalties).  

### 3. Ambiguous Cancellation Reasons  
The reasons provided for cancellation are vague. "Personal & Car related issues" doesn't give Ola any specific problem to solve.  

---

# Actionable Recommendations

Based on these findings, here are three data-driven recommendations you can propose:

## Recommendation 1: Implement a Granular Cancellation Feedback System for Drivers
**Problem:** The reason "Personal & Car related issues" is too broad.  
**Solution:** Instead of a single option, provide drivers with a multi-select checklist upon cancellation (e.g., "Flat tire," "Engine trouble," "Family emergency," "Heavy traffic").  
This would provide Ola with structured, actionable data to identify patterns. For instance, if many drivers in one area report "Heavy traffic," Ola could adjust its ETA algorithms.  

## Recommendation 2: Investigate High-Frequency Driver Cancellations  
**Problem:** The reason "The customer was coughing/sick" might be exploited by drivers to refuse rides they don't want.  
**Solution:** Create a system to flag drivers who use this reason at a rate significantly higher than the platform average.  
A follow-up automated survey could be sent to the customer in such cases ("Your driver cancelled, citing illness. Can you confirm this?").  
This can help deter drivers from misusing the feature.  

## Recommendation 3: Launch a "Successful Ride Bonus" Program  
**Problem:** The current system may not sufficiently incentivize drivers to complete rides, especially shorter or less profitable ones.  
**Solution:** Introduce a weekly bonus for drivers who maintain a high completion rate (e.g., above 95%).  
This shifts the focus from penalizing cancellations to rewarding successful service, aligning driver goals with the company's goal of reliability.
